#### Write a function to identify how many men vs. women were on board, and how many survived in each group (3 points)

```python
s=tt_df.groupby(['Sex','Survived']).size().unstack(fill_value=0)
# Group the dataset by 'Sex' and 'Survived' and count number of entries in each group
```

```python
s.columns=['Did not Survive', 'Survived'] # Rename the columns 
s['Total'] = s.sum(axis=1) #Add a new column that sums the total passenger by gender
s
```

#### 4.Write a function to identify how many men vs. women were traveling with families of 3 or more members, 
#### and how many were traveling alone or in pairs (3 points)

```python
# Make FamilySize column-number of family members onboard, including self
tt_df['FamilySize'] = tt_df['SibSp'] + tt_df['Parch'] + 1

# Adding new column TravelGroup
tt_df['TravelGroup'] = tt_df['FamilySize'].apply(
    lambda x: "Large (3+)" if x >= 3 else "Small (1-2)"
)

# Using groupby and size
result = tt_df.groupby(['TravelGroup', 'Sex']).size()

result
```

#### 6.Write a function to identify how many passengers were in each class (can be inferred from the cabin variable) (1 point)



```python
#Count of how many passengers are in each ticket class (1st, 2nd, 3rd)
tt_df.groupby(['Pclass']) ['PassengerId'].count()
```

#### 7.Write a function to identify how many passengers paid high vs. low fare tickets (1 point)

```python
median_fare=tt_df['Fare'].median() # find the median
median_fare
```

```python
#Create a new column 'fare_level'
tt_df['fare_level']=tt_df['Fare'].apply(lambda x: 'High' if x > median_fare else 'Low') 
print(tt_df['fare_level'].value_counts() )
#Count passengers by fare level
```

```python
#Save PassengerId, Name, fare_level columns to CSV
tt_df[['PassengerId', 'Name','fare_level']].to_csv('Farelevels.csv', index=False)
tt_df #dataset with new column
```

#### 8.Define a function to classify each person in an age group with the following groupings: 0-10, 11-20, 21-30, 31-40, 41-50, 51-60, 60+. Add a new column in the DataFrame which identifies each person's age group. Then, count the number of survivalists in each age group (8 points)

```python
import numpy as np
#define age group confitions
conditions=[
    (tt_df['Age']<=10),
    (tt_df['Age']<=20),
    (tt_df['Age']<=30),
    (tt_df['Age']<=40),
    (tt_df['Age']<=50),
    (tt_df['Age']<=60),
    (tt_df['Age']>60),
]
#define labels for each age group
Age_group=['0-10','11-20','21-30','31-40','41-50','51-60','60+']
tt_df['Age_group']=np.select(conditions, Age_group, default='unknown')

tt_df #df with new column Age_group
```

Or only those with larger families or those who were younger? (5 points)

```python
#Survival by Sex and Class
survivalBySexClass= tt_df.groupby(['Sex','Pclass']) ['Survived'].mean()

survivalBySexClass
```

```python
#Survival by Sex and Age Group
survivalBySexAgeGroup= tt_df.groupby(['Sex','Age_group']) ['Survived'].mean()

survivalBySexAgeGroup
```

```python
#Survival by Embarked (Port of departure)
survivalByFamily= tt_df.groupby('Embarked') ['Survived'].mean()

survivalByFamily

#Women, children, those traveling in small families, and those in 1st class.
```
